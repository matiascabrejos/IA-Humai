```python
import pandas as pd
```

```python
df_airbnb = pd.read_csv('listings.csv')
```

```python
df_price = df_airbnb.pivot_table(
    index=['neighbourhood'],
    values=['host_id'],
    aggfunc='count'
)
df_price.sort_values('host_id', ascending=False)
```

```python
df_price_3 = df_airbnb.pivot_table(
    index=['neighbourhood'],
    values=['price'],
    aggfunc='mean'
)
df_price_3.round().sort_values('price', ascending=False)
```

```python
df_group = df_airbnb.groupby('neighbourhood').aggregate({
                                        'price':['mean','median','size']})
df_group.round()
```

```python
df_price_2 = df_airbnb.pivot_table(
    index=['neighbourhood', 'name', 'room_type'],
).reset_index()
df_price_2
```
