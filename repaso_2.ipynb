```python
more_df = df[['Name', 'Age', 'Nationality', 'Club' ,'Position', 'Overall', 'Height', 'Weight', 'Special']]
```

```python
more_df
```

```python
more_df.columns
```

```python
df_filtrado = more_df.query('Overall > 90')
```

```python
df_filtrado.plot.scatter(y='Age', x='Overall',
                c='r',
                s=10
                , title='Gráfico Age/Overall' 
               ) 

plt.gca().invert_xaxis() 
plt.xlabel('Overall')
plt.ylabel('Age')
```

```python
df_filtrado.plot.hexbin(y='Age', x='Overall', gridsize=25,
              title='Gráfico Age/Overall')
plt.gca().invert_xaxis()
plt.xlabel('Overall')
plt.ylabel('Age')
```
